## Method(Mean,Median,Avg,Standerd Deviation) 
returns the mean,madian,min,max,stnader deviation (sd) of a give a dataframe.<br>
Parameters(pandas.dataframe object)<br>
returns dict

```python
def foo(d):
    rt = {}
    rt['min'] = d.min()
    rt['max'] = d.max()
    rt['mean'] = d.mean()
    rt['median'] = d.median()
    rt['sd'] = d.std()
    return rt
```

### Min Max Normalization. For Income Attribute.
    Formula v' = ((v - min)/(max - min))*(new_min-new_max)+ new_min
            new_min = 0.0 new_max = 1.0
            v' new Normalizied value
            v original Value.

```python
new_min ,new_max = 0.0,1.0
data['norm_income'] = ((data["Income"] -data["Income"].min())/(data['Income'].min()- data['Income'].max() ) * (new_min- new_max) + new_min  ) + new_min
```

```python
# Example
data['norm_income'][:4]
```

### Z-Score Normalization for Rentals Attribute.
    Formula  v' = (v- mean(of Attriute))/standerd_deviation(of attribute)
             v' = New Normalized Value
             v  = Origianl Value.

```python
data['norm_rentals'] = ((data['Rentals'] - data['Rentals'].mean())/data['Income'].std())
```
